# FIFA 22 Player Dataset - Column Explanations

This dataset contains **19,239 players** with **110 columns** of detailed FIFA 22 player data.

## üìã Column Categories and Descriptions

### **1. Basic Player Information**
- **sofifa_id**: Unique identifier for each player on SoFIFA
- **player_url**: Link to the player's page on SoFIFA
- **short_name**: Abbreviated player name (e.g., "L. Messi")
- **long_name**: Full player name (e.g., "Lionel Andr√©s Messi Cuccittini")
- **player_positions**: Positions the player can play (e.g., "RW, ST, CF")
- **age**: Player's age
- **dob**: Date of birth
- **height_cm**: Player height in centimeters
- **weight_kg**: Player weight in kilograms

---

### **2. Overall Ratings**
- **overall**: Overall player rating (0-100)
- **potential**: Maximum potential rating the player can reach
- **value_eur**: Player's market value in Euros
- **wage_eur**: Player's weekly wage in Euros
- **release_clause_eur**: Amount needed to trigger release clause

---

### **3. Club Information**
- **club_team_id**: Unique club identifier
- **club_name**: Name of the club
- **league_name**: Name of the league
- **league_level**: Level/tier of the league
- **club_position**: Position in the club
- **club_jersey_number**: Jersey number at club
- **club_loaned_from**: Club player is loaned from (if applicable)
- **club_joined**: Date player joined the club
- **club_contract_valid_until**: Year contract expires

---

### **4. National Team Information**
- **nationality_id**: Unique nationality identifier
- **nationality_name**: Player's nationality

- **nation_position**: Position in national team
- **nation_jersey_number**: Jersey number for national team

### **5. Player Attributes**

## üßπ Data Cleaning: Removing Columns with >80% Null Values

```python
# Identify columns with more than 80% null values
threshold = 0.80
null_percentages = df.isnull().sum() / len(df)

# Find columns to drop (>80% null)
columns_to_drop = null_percentages[null_percentages > threshold].index.tolist()

print("COLUMNS TO BE DROPPED (>80% NULL VALUES)")
print("="*70)
print(f"\nThreshold: {threshold*100}% null values")
print(f"Total columns to drop: {len(columns_to_drop)}")
print("\n" + "-"*70)
print(f"{'Column Name':<40} {'Null Percentage':>20}")
print("-"*70)

for col in columns_to_drop:
    percentage = (df[col].isnull().sum() / len(df)) * 100
    print(f"{col:<40} {percentage:>19.2f}%")

print("="*70)
```

```python
# Create a cleaned dataframe by dropping these columns
df_cleaned = df.drop(columns=columns_to_drop)

print("DATASET COMPARISON: BEFORE vs AFTER CLEANING")
print("="*70)
print(f"\n{'Metric':<30} {'Original':>15} {'Cleaned':>15} {'Change':>10}")
print("-"*70)
print(f"{'Number of Rows':<30} {df.shape[0]:>15,} {df_cleaned.shape[0]:>15,} {0:>10}")
print(f"{'Number of Columns':<30} {df.shape[1]:>15} {df_cleaned.shape[1]:>15} {df_cleaned.shape[1] - df.shape[1]:>10}")
print(f"{'Total Null Values':<30} {df.isnull().sum().sum():>15,} {df_cleaned.isnull().sum().sum():>15,} {df_cleaned.isnull().sum().sum() - df.isnull().sum().sum():>10,}")
print(f"{'Columns with Nulls':<30} {(df.isnull().sum() > 0).sum():>15} {(df_cleaned.isnull().sum() > 0).sum():>15} {(df_cleaned.isnull().sum() > 0).sum() - (df.isnull().sum() > 0).sum():>10}")
print("="*70)

print(f"\n‚úì Cleaned dataframe created: {df_cleaned.shape[0]:,} rows √ó {df_cleaned.shape[1]} columns")
```

```python
# Save the cleaned dataframe to a new CSV file
output_filename = 'players_22_cleaned.csv'
df_cleaned.to_csv(output_filename, index=False)

print("‚úÖ NEW CSV FILE CREATED SUCCESSFULLY!")
print("="*70)
print(f"\nFile name: {output_filename}")
print(f"Location: d:\\Git Repo\\Data-Visualization-Lab\\Project\\{output_filename}")
print(f"\nDataset shape: {df_cleaned.shape[0]:,} rows √ó {df_cleaned.shape[1]} columns")
print(f"File size: {df_cleaned.memory_usage(deep=True).sum() / (1024*1024):.2f} MB (estimated)")
print("\n" + "="*70)

# Display first few rows of cleaned data
print("\nPreview of cleaned data:")
df_cleaned.head()
```
